```python
import cv2
from pathlib import Path

def extract_frames(video_path, output_folder, frame_skip=5, prefix="frame"):
    video_path = Path(video_path)
    output_folder = Path(output_folder)

    if not video_path.exists():
        raise FileNotFoundError(f"Video tidak ditemukan: {video_path.resolve()}")
    if frame_skip < 1:
        raise ValueError("frame_skip harus >= 1")

    output_folder.mkdir(parents=True, exist_ok=True)

    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise RuntimeError(f"Gagal membuka video: {video_path}")

    frame_id = 0
    saved_id = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_id % frame_skip == 0:
            filename = output_folder / f"{prefix}_{saved_id:06d}.jpg"
            cv2.imwrite(str(filename), frame)
            saved_id += 1

        frame_id += 1

    cap.release()
    print(f"Done: {saved_id} frames saved to {output_folder.resolve()}")
    return saved_id
```

```python
video_path = Path("video_datasets/tank_video_1.mp4")
output_dir = Path("video_datasets/frames/tank_video_1_skip10")

saved_count = extract_frames(video_path, output_dir, frame_skip=10, prefix="tank1")
print("Saved frames:", saved_count)
```
